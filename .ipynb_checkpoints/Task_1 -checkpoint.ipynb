## So, What is Optical Character Recognition or, OCR ?
<b> Optical Character Recognition or Optical Character Reader is the electronic or mechanical conversion of images of typed, handwritten or printed text into machine-encoded text, whether from a scanned document, a photo of a document, a scene-photo or from subtitle text superimposed on an image.

- With **Python** we can do OCR by using either " easyocr " or, " tessearct " module.
- But, I am going to be using the easyocr module for this particular task.

### First Step : Importing all the dependencies. 

```python
import easyocr
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

### Second Step : Reading an Image 
Here a single line image namely, **"test.jfif"** is being scanned using a pre-trained model called **"Reader"**, from that model we use a fucntion/method called **readtext()**, which scans for the text and gives us the output in the from of a list containing 3 elements. 
- The first element consists of the coordinates of the Text.
- The second element gives us the Text present on the Image.
- The third element gives us the matching accuracy of the Text.

```python
Img1 = 'test.jfif'

reader = easyocr.Reader(['en'], gpu = False)
result = reader.readtext(Img1)
print(result)
```

- So we can see from above that the text is **'STOP'**, with a matching accuracy of **0.99192...** or **99.19 %**.
- The co-ordinates are also given in form of a list.

### Third Step : Drawing the results.
Now, from the object : **result** above, I will be finding the **top-left** & **lower-right** co-ordinates, then using them to make a rectangle around the text using rectangle function/method from openCV (cv2) and then showing the output using pyplot module.

```python
coord = result[0][0]   # Co-ordinates
text = result[0][1]    # Text in the Image

# Taking the bounding values for creating a rectangle around the text.

x_min, y_min = [min(idx) for idx in zip(*coord)]    # Top-Left (x,y) coordinates
x_max, y_max = [max(idx) for idx in zip(*coord)]    # Bottom-Right (x,y) coordinates

image = cv2.imread(Img1)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.rectangle(image, (x_min, y_min), (x_max,y_max), (0,255,0), 2)

# Showing the text in the image above the rectangle.
image = cv2.putText(image, text, (x_min+30, y_min), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

plt.imshow(image)
plt.show()
```

### For Images containing multiple lines
- Here, we will be seeing the output for an image namely **"test 2.jpg"**, which has multiple lines. So, let's check the output for it.

```python
# For handling Images containing multiple lines
Img2 = 'test 2.jpg'
image = cv2.imread(Img2)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

reader1 = easyocr.Reader(['en'], gpu = False)
result1 = reader1.readtext(Img2)

plt.imshow(image)
print(result1)
```

**So, upto this we have handled single line image and as well as multi-line image.** 
- Let us take another multi-line image **(test final.png)** and find the various texts and print them simultaneouly on their boundaries using a for-loop.

```python
# For Handling Multiple lines using loop.
Img3 = 'test final.png'
image = cv2.imread(Img3)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

reader = easyocr.Reader(['en'], gpu = False)
outputList = reader.readtext(Img3)

# Using loop for traversal in outputList
for detection in outputList:
    coord = detection[0]
    text = detection[1]
    
    x_min, y_min = [min(idx) for idx in zip(*coord)]
    x_max, y_max = [max(idx) for idx in zip(*coord)]
    
    image = cv2.rectangle(image, (x_min, y_min), (x_max,y_max), (255,0,10), 5)
    image = cv2.putText(image, text, (x_min+30, y_min), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0,0,0), 5)
    
plt.figure(figsize = (15, 15))
plt.imshow(image)
plt.show()
```
